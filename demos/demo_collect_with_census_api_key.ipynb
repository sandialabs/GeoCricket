# Full Collect Demo - with Census API Key

Example of how to collect information from an area.

Census API key utlized (and required) for census data collection.

```python
import warnings
warnings.filterwarnings("ignore")

import geocricket as gc
import pandas as pd
from pathlib import Path

```

Load Census API key.

This is unique to this situation - api_key is simple string

```python
census_key_path = Path(r"..\..\00-api_keys\census_api_key.csv")
temp_df = pd.read_csv(census_key_path)
api_key = temp_df['key'].iloc[0]
```

Define input and output paths

```python
base_path = Path(r".\demo_data")

result_path = base_path / 'results'

geometry_fp = base_path / 'abq_blob.gpkg'
```

Collect data with a census api key

```python
collected = gc.collect(
    geometry_fp,
    result_path,
    census_api_key=api_key,
)
```
