# Full Collect Demo - without Census API Key

Example of how to collect information from an area without a census api key.

Census geometry will still be collected, but statistics will not.

```python
import warnings
warnings.filterwarnings("ignore")

import geocricket as gc
from pathlib import Path
```

Define input and output paths

```python
base_path = Path(r".\demo_data")

result_path = base_path / 'results'

geometry_fp = base_path / 'abq_blob.gpkg'
```

Collect data with a census api key

```python
collected = gc.collect(
    geometry_fp,
    result_path,
)
```

Display collected results
